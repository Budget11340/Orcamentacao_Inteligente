```python
import pandas as pd

# Definindo o caminho do arquivo de entrada e de saída
input_file = 'Processed_Normalized_Data.csv'
output_file = 'descritive_statistics.csv'

# Carregando os dados
df = pd.read_csv(input_file)

# Verificando se a coluna 'Year' existe antes de tentar removê-la
if 'Year' in df.columns:
    df = df.drop(columns=['Year'])

# Calculando estatísticas descritivas
desc_stats = df.describe().transpose()
desc_stats['kurtosis'] = df.kurtosis()
desc_stats['skewness'] = df.skew()

# Renomeando as colunas para garantir consistência
desc_stats = desc_stats.rename(columns={
    'count': 'count',
    'mean': 'mean',
    'std': 'std',
    'min': 'min',
    '25%': '25%',
    '50%': '50%',
    '75%': '75%',
    'max': 'max',
    'kurtosis': 'kurtosis',
    'skewness': 'skewness'
})

# Salvando as estatísticas descritivas em um novo arquivo CSV
desc_stats.to_csv(output_file)

# Exibindo as estatísticas descritivas no notebook
desc_stats
```
